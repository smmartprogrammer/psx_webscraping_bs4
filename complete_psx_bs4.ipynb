```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.psx.com.pk/market-summary/'
request = requests.get(url)
request
```

```python
soup = BeautifulSoup(request.content, 'html.parser')
```

```python
table_div = soup.find('div', class_='automobile-div')
table_div
```

```python
rows = table_div.find_all('tr')
```

```python
all_data = []

for i in rows:
    table_data = i.find_all(['td', 'th'])  # 'th' for header, 'td' for body
    row = [td.text.strip() for td in table_data]
    if row:  # avoid empty rows
        all_data.append(row)

# Print all rows after loop
for row in all_data:
    print(row)
```

```python
for row in all_data:
    print(len(row), row)
```

```python
import csv

# Save all_data to CSV file
with open("psx_market_summary.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerows(all_data)

print("CSV file saved as psx_market_summary.csv")
```
